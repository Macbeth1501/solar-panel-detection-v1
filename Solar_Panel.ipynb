```python
import yaml

# Specify the correct path to your YAML file
file_path = r"C:\Users\Rochan\Desktop\Coding\Micro_Project_2\data.yaml"

# Read the YAML file
with open(file_path, 'r') as stream:
    data = yaml.safe_load(stream)
    #num_classes = data['nc']  # Extract the number of classes
    num_classes=1;
# Print the number of classes
print(f"Number of classes: {num_classes}")
```

```python
import os

# Define a project directory for all results
project = r"C:\Users\Rochan\Desktop\Coding\Micro_Project_2\result"

# Define a subdirectory for this specific training
name = "300_epochs-"  # Note that if you run the training again, it creates a directory: 200_epochs-2

# Combine the project directory and subdirectory
results_dir = os.path.join(project, name)

# Ensure the directory exists
os.makedirs(results_dir, exist_ok=True)

print("Results will be saved to:", results_dir)
```

```python
# Train the model
results = model.train(data=r"C:\Users\Rochan\Desktop\Coding\Micro_Project_2\data.yaml",
                      project=project,
                      name=name,
                      epochs=300,
                      patience=0, # Setting patience=0 to disable early stopping
                      #batch=4,
                      imgsz=640
                      )
```
